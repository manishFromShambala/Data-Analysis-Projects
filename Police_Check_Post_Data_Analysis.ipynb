```python
#For Speeding, were men or women stopped more often
if data[data["driver_gender"]=="M"].violation[data["violation"]=="Speeding"].value_counts().sum() > data[data["driver_gender"]=="F"].violation[data["violation"]=="Speeding"].value_counts().sum():
  print("Men were stopped more often")
else:
  print("Women were stopped more ofter")
```

```python
# Does gender impacts who gets stopped(search_conducted) during a stop
data.groupby('driver_gender').search_conducted.sum()
```
